```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot")

plt.plot(x,y)
```

```python
x = [1,51,52,60,80,98]
y = [11,154,210,276,309,348]

plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth=2,markersize=9)
plt.grid(True)
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1) # 1 row, 2 columns, right now I am ploting 1st graph
plt.plot(x,y1,color="green")
plt.title("Plot 1")


plt.subplot(1,2,2) # 1 row, 2 columns, right now I am ploting 2nd graph
plt.plot(x,y2,color="red")
plt.title("Plot 2")
```

```python
# Bar plot

categories = ["A","B","C","D","E"]
values = [5,7,3,8,6]

plt.bar(categories,values,color="purple")
```

```python
# Histogram
data = [1,5,47,14,8,8,1,5,2,4,4,7,9,2,5,4,]
plt.hist(data,bins=5,edgecolor="black")
```

```python
# Scatter plot

x = [1,2,4,2,8]
y = [10,5,11,15,74]

plt.scatter(x,y,marker="x")
```

```python
#Pie chart
labels = ["A","B","C","D"]
sizes = [30,20,40,10]
colors = [ "gold","yellowgreen","lightcoral","lightskyblue"]
explode = [0.1,0,0,0]

plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",explode=explode)
```

```python

import pandas as pd

sales_data = pd.read_csv("supermarket_sales.csv")
sales_data.info()
```

```python
total_sales_by_city = sales_data.groupby("City")["Total"].sum()
total_sales_by_city
```

```python

total_sales_by_city.plot(kind="bar",color="teal",x="City",y="Total sales")
```

```python
sales_trend = sales_data.groupby("Date")["Total"].sum()
sales_trend
```

```python
sales_trend.plot(kind="line",color="teal",x="City",y="Total sales")
```
