```python
import pandas as pd
df = pd.read_csv("supermarket_sales.csv")
df.head()
```

```python
df.describe()
```

```python
# Handling missing values
df.isnull().any(axis=0)
```

```python
df_filled = df.fillna(0) # All null values will be replaced by 0 -> Bad practice
df["gross_income_fillNA"] = df["gross income"].fillna(df["gross income"].mean())
df.head()
```

```python
# Renaming columns

df = df.rename(columns={"Date":"Sale Date"})
df.head()
```

```python
df["New Quantity"] = df["Quantity"].apply(lambda x:x*2)
df.head()
```

```python
# Data aggregating and grouping
group_mean = df.groupby("Branch")["Quantity"].mean()
print("Group by single:" , group_mean,"\n\n")
group_mean = df.groupby(["Branch","Gender"])["Quantity"].mean()
print("Group by multiple:" , group_mean)
```

```python
# Merging and joining

df1 = pd.DataFrame({"Key":["A","B","C"],"Value1":[1,2,3]})
df2 = pd.DataFrame({"Key":["A","B","D"],"Value2":[4,8,10]})

print("Merge: \n",pd.merge(df1,df2,on="Key",how="outer")) # Same as SQL joins
```
